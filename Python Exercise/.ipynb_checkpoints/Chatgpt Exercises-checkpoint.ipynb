2) **Word Frequency Counter**: Build a program that reads a text file and counts the frequency of each word.

**Grade: 5.5/10***

**Feedback**
I would give your attempt a 5.5 out of 10. You've tackled the core idea of counting word frequencies, but there are areas for improvement in terms of efficiency and handling special cases.

1) **Efficiency**: Your current approach is a bit inefficient as it involves iterating through the entire text multiple times. You can achieve better efficiency by splitting the text into words and using a dictionary to count the occurrences in a single pass.

2) **Case Sensitivity**: Your current implementation treats words with different cases (e.g., "Hello" and "hello") as distinct words. Consider converting all words to lowercase to ensure accurate frequency counts.

3) **Punctuation**: Your implementation doesn't handle punctuation marks well. Words like "Hello," and "Hello" would be treated as different words. You can preprocess the text to remove punctuation before counting.

4) **Function Name**: The function name count_of_each_distinct is a bit long and could be more concise while conveying its purpose.

```python
# my answer

# example path: C:\Users\Ahmed\Desktop\Learning\Python Exercise\Random Text.txt

# input file path
path = input("Please enter full file path (without quotations or double backslashes)")

# retrieve count of each distinct word
def count_of_each_distinct(x):
    final_list = {}
    counter = 0
    
    for i in set(x.split()):
        for j in x.split():
            if i == j:
                counter += 1
        final_list[i] = counter
        counter = 0
    return final_list

# read file    
with open(path, 'r', encoding='UTF-8') as f:
    k = f.read()

count_of_each_distinct(k)
```

```python
# model answer

import string

def count_word_frequencies(text):
    word_count = {}
    words = text.lower().translate(str.maketrans('', '', string.punctuation)).split()

    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    return word_count

# Input file path
path = input("Please enter the full file path (without quotations or double backslashes): ")

# Read file
with open(path, 'r', encoding='UTF-8') as f:
    content = f.read()

frequency_count = count_word_frequencies(content)
for word, count in frequency_count.items():
    print(f"'{word}' appears {count} times.")
```

```python
#model answer

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load data
data = pd.read_csv(r"C:\Users\Ahmed\Desktop\Learning\Python Exercise\titanic.csv")

# Basic data exploration
print(data.info())
print(data.describe())

# Analyze Age Distribution and Survival
data['Age Group'] = pd.cut(data['Age'], bins=[0, 10, 20, 30, 40, 50, 60, 100])
age_grp_sur = pd.crosstab(data['Survived'], data['Age Group'], normalize='columns')

# Analyze Gender Distribution and Survival
sex_prob = pd.crosstab(data['Survived'], data['Sex'], normalize='columns')

# Analyze Pclass Distribution and Survival
class_prob = pd.crosstab(data['Survived'], data['Pclass'], normalize='columns')

# Visualization
plt.figure(figsize=(10, 6))
sns.heatmap(age_grp_sur, cmap='Blues', annot=True)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survived')

plt.figure(figsize=(8, 6))
sex_prob.plot(kind='bar')
plt.title('Survival Rate by Gender')
plt.xlabel('Survived')
plt.ylabel('Percentage')
plt.xticks(rotation=0)

print(class_prob)
# Class 1 has the highest survival rate, followed by Class 2 and Class 3.

plt.show()
```

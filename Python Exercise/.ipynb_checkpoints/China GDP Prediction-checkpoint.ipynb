```python
# Import required libraries
import pandas as pd
import prophet
import matplotlib.pyplot as plt


# Read in the GDP data
data = pd.read_csv('C:\\Users\\Ahmed\\Desktop\\Learning\\Datasets\\China GDP by Province\\Chinas GDP in Province En.csv')

data.head()
```

```python
# Cleaning Data
data.rename(columns={'Unnamed: 0': 'Year'}, inplace=True)

data.set_index('Year', inplace=True)

data = data.reset_index()  # step 1
data_melt = pd.melt(data, id_vars=["Year"], var_name="Province", value_name="GDP")  # step 2
data_melt = data_melt.rename(columns={"Year": "ds", "GDP": "y"})  # step 3
data_melt = data_melt[data_melt.Province != 'level_0']
data_melt = data_melt[data_melt.Province != 'index']
data_melt
```

```python
# Fit the model to the data
model = prophet.Prophet()
model.fit(data_melt)

# Generate future dates to forecast
future_dates = model.make_future_dataframe(periods=30, freq='Y')

# Make the forecast
forecast = model.predict(future_dates)

# Plot the forecast
model.plot(forecast, xlabel='Year', ylabel='GDP')
plt.title('GDP Forecast for China Provinces')
plt.show()
```

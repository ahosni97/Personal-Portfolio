```python
# Import Data DONE

import pandas as pd
Amazon = pd.read_csv(r"C:\Users\Ahmed\Desktop\Learning\Python Exercise\Amazon\amazon - Original.csv", index_col=0)


Amazon['discounted_price'] = Amazon['discounted_price'].str.replace('₹','')
Amazon['actual_price'] = Amazon['actual_price'].str.replace('₹','')
Amazon['discount_percentage'] = Amazon['discount_percentage'].str.replace('%','')

# Replacing all commas with semicolon not to mess up save as csv
Amazon['discounted_price'] = Amazon['discounted_price'].str.replace(',','')
Amazon['actual_price'] = Amazon['actual_price'].str.replace(',','')
Amazon['discount_percentage'] = Amazon['discount_percentage'].str.replace(',','')
Amazon['rating_count'] = Amazon['rating_count'].str.replace(',','')
for col in Amazon:
    Amazon[col] = Amazon[col].str.replace(',',';')

# Replacing or removing characters in strings (eg. $) DONE
#Amazon[['discounted_price', 'actual_price']] = Amazon[['discounted_price', 'actual_price']].str.replace('₹','')
#Amazon['discount_percentage'] = Amazon['discount_percentage'].str.replace('%','')
#Amazon[['discounted_price', 'actual_price', 'discount_percentage', 'rating_count']] = Amazon[['discounted_price', 'actual_price', 'discount_percentage', 'rating_count']].str.replace(',','')

# Fixing some entries
Amazon['rating'].iloc[1279] = 0


# Expanding category col to 6 columns and reordering df
Amazon[['category', 'sub_category1', 'sub_category2', 'sub_category3', 'sub_category4', 'sub_category5', 'sub_category6']] = Amazon['category'].str.split(pat="|", expand=True)
Amazon = Amazon[['product_name',
 'category',
 'sub_category1',
 'sub_category2',
 'sub_category3',
 'sub_category4',
 'sub_category5',
 'sub_category6',
 'discounted_price',
 'actual_price',
 'discount_percentage',
 'rating',
 'rating_count',
 'about_product',
 'user_id',
 'user_name',
 'review_id',
 'review_title',
 'review_content',
 'img_link',
 'product_link'
 ]]


    
# Converting columns types to int, float or str
Amazon[["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]] = Amazon[["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]].apply(pd.to_numeric)
Amazon[['product_name', 'about_product', 'user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link']] = Amazon[['product_name', 'about_product', 'user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link']].astype('|S', errors='ignore')

# Finding and replacing NA
Amazon.isna().sum()
for col in Amazon:
    if Amazon[col].dtype == 'O':
        Amazon[col].fillna('-', inplace=True)
    else:
        Amazon[col].fillna(0, inplace=True)
"""        
Amazon['rating_count'].fillna(0, inplace=True)
Amazon['sub_category2'].fillna('-', inplace=True)
Amazon['sub_category3'].fillna('-', inplace=True)
Amazon['sub_category4'].fillna('-', inplace=True)
Amazon['sub_category5'].fillna('-', inplace=True)
Amazon['sub_category6'].fillna('-', inplace=True)
"""

# Amazon.iloc[51, 1]

# Removing duplicate product_id
Amazon.reset_index(inplace=True)
Amazon.drop_duplicates(subset='product_id', inplace=True)
Amazon.set_index('product_id', inplace=True)

# Random logical checks
(Amazon['discounted_price'] - Amazon['actual_price'] < 0).values
(Amazon['rating_count'] < 0).values.any()
(Amazon['product_link'].str.startswith('http') == False).values.any()

# Checking and setting max char len for each column (max 1000)
for col in Amazon:
    if Amazon[col].dtype == 'O':
        print(col, ": ", Amazon[col].str.len().max())
Amazon['about_product'] = Amazon['about_product'].str.slice(0, 1000)
Amazon['review_content'] = Amazon['review_content'].str.slice(0, 1000)

# Exporting to csv file
#Amazon.to_csv('Amazon_PYClean_NoCategory.csv')
Amazon.to_excel('Amazon_PYClean_NoComma_Excel_NoDuplicates.xlsx')
```
